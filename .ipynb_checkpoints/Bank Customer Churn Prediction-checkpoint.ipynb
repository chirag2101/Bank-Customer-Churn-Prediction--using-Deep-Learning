```python
#Get unique values from each object columns
def print_unique_col_values(df):
    for col in df:
        if df[col].dtypes=='object':
            print(f'{col}:{df[col].unique()}')
```

```python
print_unique_col_values(df)
```

```python
#Replace Female with 0 and Male with 1 
df['Gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
# 0=>Female , 1=>Male
df['Gender'].value_counts()
```

```python
#Generate dummies columns for Geography and Card Type
df2=pd.get_dummies(data=df,columns=['Geography','Card Type'])
```

```python
df2
```

```python
df2.dtypes
```

```python
#Convert boolean values into integer
columns_to_convert=['Geography_France','Geography_Germany','Geography_Spain','Card Type_DIAMOND','Card Type_GOLD','Card Type_PLATINUM','Card Type_SILVER']
df2[columns_to_convert]=df2[columns_to_convert].astype(bool)
df2[columns_to_convert]=df2[columns_to_convert].astype(int)
```

```python
df2.dtypes
```

```python
df2.sample(3)
```

```python
#Scaling Process
cols_to_scale=['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary','Satisfaction Score','Point Earned']

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(3)
```

```python
#Define X and y
X=df2.drop(['Exited'],axis='columns')
y=df2['Exited']
```

```python
#Split training and testing data using train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=5)
```

```python
len(X_train),len(X_test)
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
#Model builing using TensorFlow and Keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(19,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train,y_train,epochs=100)
```

```python
# Model accuracy on test data
model.evaluate(X_test,y_test)
```

```python
#Prediction
yp=model.predict(X_test)
yp[:5]
```

```python
y_pred=[]
for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
#Predicted Values
y_pred[:10]
```

```python
#True values
y_test[:10]
```

```python
#Classification Report
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
